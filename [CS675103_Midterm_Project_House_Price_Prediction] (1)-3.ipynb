### Dealing with missing data

We used 2 method to deal with the missing values which are present in the dataset.

<b>1.Drop Missing values:</b>

        If a feature contains more that 50% of the missing value then it better to drop that column as it will have very less or negetive impact on our model accuracy. 

<b>2.Filling Missing values:</b>

        In this project we are taking the most frequent value of the feature and we are filling that value wherever we have msiing value  

```python
dropcol=[] #Dropping the columns having missing values more than 500
for col in train.columns:
    if train[col].isna().sum()>500:
        dropcol.append(col)
print(dropcol)
```
